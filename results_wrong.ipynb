```python
#配置环境变量
import chess
import chess.pgn
from stockfish import Stockfish
stockfish=Stockfish('C:/Users/Administrator/Desktop/stockfish-windows-x86-64-avx2/stockfish/stockfish-windows-x86-64-avx2.exe')
pgn_file=open(r'C:\Users\Administrator\Desktop\2023\Nepomniachtchi, Ian_vs_Ding, Liren_2023.04.09.pgn')
game=chess.pgn.read_game(pgn_file)
```

```python
out_dict = {}
out_dict['Black'] = game.headers['Black']
out_dict['Black'] = game.headers['Black']
```

```python
board=chess.Board()#初始化棋盘
move_scores=[] #存每一步棋子的分数
stockfish.set_depth(20)
```

```python
for move in game.mainline_moves():
    board.push(move)#更新前面空的borad的状态
    stockfish.set_position(board.fen())#新的棋盘状态传给stockfish以便进行评估
    evaluation=stockfish.get_evaluation()#拿到stockfish给的分数
    if isinstance(evaluation, dict):
        score = evaluation['value']  # 获取评分数值
    else:
        score = evaluation  # 如果直接返回数值，直接赋值
    move_scores.append((move,score))#生成一个类似坐标tuple，用于画点
```

```python
print(move_scores)
```

```python
for move,score in move_scores:
    print(f"Move: {move}, Score: {score}")
```

```python
move_numbers = [i + 1 for i in range(len(move_scores))]  # x轴：步数，i + 1是为了从1开始
scores = [score for _, score in move_scores]  # y轴：评分，提取评分部分
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))  # 设置图像大小
plt.plot(move_numbers, scores, label='Stockfish Evaluation', color='b')  # 绘制评分曲线
plt.title("Evaluation of Each Move")  # 图表标题
plt.xlabel("Move Number")  # x轴标签
plt.ylabel("Stockfish Score")  # y轴标签

plt.legend()  # 显示图例
plt.grid(True)  # 显示网格
plt.show()  # 显示图表
```
